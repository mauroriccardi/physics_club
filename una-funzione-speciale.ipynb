## Una funzione speciale

Possiamo definire una nuova funzione usando una equazione differenziale piuttosto comune:

$$
y'(x) = y(x)
$$

Per avere una soluzione unica, però, come sappiamo, dobbiamo porre un problema di Cauchy. Quindi definiamo la nuova funzione (derivabile) $y = \exp(x)$ come la funzione che risolve il seguente problema di Cauchy:

$$
\left\{
\begin{array}{rcl}
\exp'(x) & = & \exp(x) \\
\exp(0) & = & 1 \\
\end{array}
\right.
$$

Nel seguito per brevità scriveremo $e(x) = \exp(x)$, visto che i passaggi che faremo saranno già abbastanza pesanti...

### Proprietà della funzione $e(x)$

Ovviamente la funzione ha le proprietà di avere valore $1$ per $x=0$, e di essere uguale alla sua derivata ovunque (che sono le richieste del problema di Cauchy). Ma questa funzione (a proposito, si chiama **esponenziale**) ha anche altre proprietà che possiamo ricavare: solo che ci vorrà un po' di impegno.

Anzitutto usiamo la equazione differenziale per ricavare una diversa espressione della funzione. Ragioneremo in maniera euristica e poco formale, come al solito: ma sperabilmente impareremo qualcosa di utile in generale sulle equazioni differenziali.

Diciamo che vogliamo calcolare il valore della funzione $e(x)$ per un certo valore $X$ della variabile indipendente $x$, e dividiamo l'intervallo compreso tra $0$ e $x_1$ in $N$ piccoli intervalli tutti uguali:

$$
0 = x_0 < x_1 < x_2 < \dots < x_N = X \qquad x_k = x_{k-1} + \Delta x = x_{k-1} + \frac{X}{N}
$$

cioè abbiamo diviso l'intervallo $\left[0,X\right]$ in $N$ intervalli di ampiezza uguale $\Delta x = \frac{X}{N}$. Possiamo calcolare numericamente in maniera approssimata i valori della funzione $e(x)$ per $x=x_k$, partendo dal primo $x_0$:

$$
e(x_0) = e(0) = 1 \ , \ e(x_1) = e(x_0+\Delta x) \approx e(x_0) + e'(x_0) \Delta x \ , \ e(x_2) = e(x_1+\Delta x) \approx e(x_1) + e'(x_1)\Delta x
$$

Usando il fatto che $e'(x) = e(x)$ abbiamo per ogni punto $x_k$

$$
e(x_k) \approx e(x_{k-1}) + e'(x_{k-1})\Delta x = e(x_{k-1}) + e(x_{k-1})\Delta x = e(x_{k-1})\left(1+\Delta x\right) = e(x_{k-1})\left(1+\frac{X}{N}\right)
$$

Questi termini sono quelli di una progressione geometrica di ragione $\left(1+\frac{X}{N}\right)$, quindi il termine generale è:

$$
e(x_k) = e(x_0)\cdot\left(1+\frac{X}{N}\right)^N = e(0)\cdot\left(1+\frac{X}{N}\right)^N = \left(1+\frac{X}{N}\right)^N
$$

Adesso, con un po' di coraggio, dobbiamo calcolare la potenza del binomio per un dato valore di $N$ per calcolare il valore di $e(x)$ in $x=X$ (e soprattutto, se avete un matematico per casa, chiedetegli di farsi un passeggiata di una mezz'oretta, purché non sia presente...):

$$
\begin{array}{rl}
e(X) = \left(1+\frac{X}{N}\right)^N = 1 + N \frac{X}{N} & + \frac{N(N-1)}{2} \frac{X^2}{N^2} + \dots + \\
& + \frac{N(N-1)(N-2)\cdot\dots\cdot(N-k+1)}{2\cdot 3\cdot\dots \cdot k} \frac{X^k}{N^k} + \dots + \\
& + \frac{N(N-1)(N-2)\cdot\dots\cdot 1}{2\cdot 3\cdot\dots \cdot N} \frac{X^N}{N^N}
\end{array}
$$

questa somma contiene tanti termini per quanto vale $N$. Adesso immaginiamo di prendere un valore di $N$ molto grande, allora la somma diventa una sommatoria di una grande quantità di termini, e il termine generale ha la forma:

$$
\frac{N(N-1)(N-2)\cdot\dots\cdot(N-k+1)}{2\cdot 3\cdot\dots \cdot k} \frac{X^k}{N^k} = \frac{N}{N}\frac{N-1}{N}\frac{N-2}{N}\cdot\dots\cdot\frac{N-k+1}{N}\frac{X^k}{2\cdot 3\cdot\dots \cdot k}
$$

Abbiamo riorganizzato i fattori. I fattori $\frac{N}{N}$, $\frac{N-1}{N}$, $\frac{N-2}{N}$ e così via, quando $N$ diventa molto grande in realtà spariscono, infatti:

$$
\frac{N}{N}\frac{N-1}{N}\frac{N-2}{N}\cdot\dots\cdot\frac{N-k+1}{N} = 1 \cdot \left(1-\frac{1}{N}\right) \cdot \left(1-\frac{2}{N}\right) \cdot \left(1-\frac{3}{N}\right) \cdot \dots \cdot \left(1-\frac{k-1}{N}\right)
$$
e vediamo che quando $N$ diventa molto grande (cioè _tende all'infinito_ $N\rightarrow\infty$) tutti questi fattori tendono a $1$. Quindi otteniamo:

$$
\frac{N(N-1)(N-2)\cdot\dots\cdot(N-k+1)}{2\cdot 3\cdot\dots \cdot k} \frac{X^k}{N^k} \rightarrow \frac{X^k}{2\cdot 3\cdot\dots \cdot k} = \frac{X^k}{k!}
$$
con la notazione del **fattoriale** $k! = 1 \cdot 2 \cdot 3 \cdot \dots \cdot k$. Quindi, per un numero molto grande di intervalli $N$

$$
e(X) \approx \left(1+\frac{X}{N}\right)^N \rightarrow 1 + \frac{X}{1!} + \frac{X^2}{2!} + \frac{X^3}{3!} + \dots + \frac{X^k}{k!} + \dots
$$

Per avere l'uguaglianza _esatta_ dobbiamo prendere un numero _infinito di termini_ della somma a destra!
L'avevo detto che ci voleva un po' di coraggio!

L'espressione $e(x) \approx \left(1+\frac{x}{N} \right)^N$ per $N$ _molto grande_ permette di calcolare comunque la funzione esponenziale con una certa precisione. Se non fosse che le moderne calcolatrici scientifiche hanno un apposito tasto per calcolare la stessa cosa in un colpo solo, intendo...

C'è una proprietà che possiamo dedurre da questa, anche se non è immediata, anche se manteniamo il modesto livello di rigore tenuto fin qui. La proprietà in questione è questa:

$$
e(X)e(Y) = e(X+Y)
$$

Usiamo quanto abbiamo trovato prima, dividiamo i due intervalli $\left[0,X\right]$ e $\left[0,Y\right]$ nello stesso numero $N$ di intervallini, e poi immaginiamo di prendere per $N$ un numero veramente molto grande (come prima).

$$
e(X)e(Y) \approx \left(1+\frac{X}{N}\right)^N \left(1+\frac{Y}{N}\right)^N = \\
\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad = \left[\left(1+\frac{X}{N}\right)\left(1+\frac{Y}{N}\right)\right]^N = \\
\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad = \left[1+\frac{X+Y}{N}+\frac{X\cdot Y}{N^2}\right]^N
$$

Adesso bisogna dimostrare che dei tre termini tra parentesi solo i primi due contano per $N$ molto grande ($N\rightarrow\infty$)...

Usiamo una incognita ausiliaria, per comodità: $A = X + Y + \frac{XY}{N}$, quindi dobbiamo calcolare:

$$
\begin{array}{rl}
\left(1+\frac{A}{N}\right)^N = 1 + N \frac{A}{N} & + \frac{N(N-1)}{2} \frac{A^2}{N^2} + \dots + \\
& + \frac{N(N-1)(N-2)\cdot\dots\cdot(N-k+1)}{2\cdot 3\cdot\dots \cdot k} \frac{A^k}{N^k} + \dots + \\
& + \frac{N(N-1)(N-2)\cdot\dots\cdot 1}{2\cdot 3\cdot\dots \cdot N} \frac{A^N}{N^N}
\end{array}
$$

e come prima otteniamo una somma di termini di questa forma

$$
\frac{N(N-1)(N-2)\cdot\dots\cdot(N-k+1)}{2\cdot 3\cdot\dots \cdot k} \frac{A^k}{N^k} = \\
\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad = 1 \cdot \left(1-\frac{1}{N}\right) \cdot \left(1-\frac{2}{N}\right) \cdot \left(1-\frac{3}{N}\right) \cdot \\
\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad \cdot \dots \cdot \left(1-\frac{k-1}{N}\right) \frac{\left(X+Y+\frac{XY}{N}\right)^k}{k!} 
$$

Quando $N\rightarrow\infty$ come prima i primi fattori tendono a $1$, mentre nel secondo fattore

$$
\frac{\left(X+Y+\frac{XY}{N}\right)^k}{k!} \rightarrow \frac{\left(X+Y\right)^k}{k!}
$$

Cioè otteniamo

$$
e(X)e(Y) = 1 + \frac{X+Y}{1} + \frac{(X+Y)^2}{2!} + \frac{(X+Y)^3}{3!} + \dots + \frac{(X+Y)^k}{k!} + \dots = e(X+Y)
$$

laddove l'ultima uguaglianza viene ricavata per confronto diretto.

>Adesso abbiamo finito, e abbiamo fatto vedere che
$$
e(X) \cdot e(Y) = e(X+Y)
$$

### Altra lettura delle proprietà dell'esponenziale

Le due proprietà

$$
\left\{
\begin{array}{rcl}
e(X) \cdot e(Y) & = & e(X+Y)\\
e(0) & = & 1 \\
\end{array}
\right.
$$

sono le stesse proprietà della moltiplicazione di potenze con la stessa base! Infatti si può mostrare che esiste un numero $e$ tale che la funzione esponenziale consista nell'elevare questo numero alla potenza $x$:

$$
e(x) = e^x
$$

In questo modo semplicemente applicando le proprietà delle potenze con la stessa base possiamo ricordare con facilità le proprietà dell'esponenziale.
