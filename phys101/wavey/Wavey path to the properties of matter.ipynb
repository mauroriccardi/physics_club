## de Broglie Waves

* from _Sixty Symbols_
* running time **6:56**

In 1924, as a young student, duke Louis de Broglie, in his doctoral thesis, came out with this outrageous idea, that one could associate to any particle a wavelength, in the very same way you did with photons. Nobody in the commission panel could understand de Broglie's thesis, and that almost costed him his PhD title: luckily enough Einstein himself was called in to judge, and that toppled the needle in de Broglie's favour.

Five years after that de Broglie was awarded a Nobel prize for Physics, because he opened up a totally new field of Physics.

## Particle-Wave duality

1.  * **Part 1**
    * from _minutephysics_
    * running time **1:06**
1.  * **Part 2**
    * from _minutephysics_
    * running time **1:03**
    
Light waves can be particles, and particles can be waves at times: is this an ambiguity? No, it's a _Duality_: matter has particle like properties, and it has wave like properties too, the latter usually overcome the former when the scale at which we observe the phenomena is quite small. Point is, the two (_dual_) aspects of matter are not shown at the same time, they are _complementary_. But how can we think of a particle that is a particle, but can be a wave as well, when circumstances so require?

## What is Quantum Tunneling?

* from _minutephysics_
* running time **1:04**

The fact that the propagation of quantum particles implies that a particle like an electron has a very weird characteristic: far from being a point-like corpuscle, it is, by its very own nature, _spread_ around in space, and spread around must be at least some of the physical quantities that refer to it (like electric charge). It's not the ordinary picture we have in mind when we think about particles, let's see a notorious (and notoriously weird) consequence of this fact about matter.
